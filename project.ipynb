Creating the Bar graph for the rent


```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file
data = pd.read_csv("House_Rent_Dataset.csv")

# Create a DataFrame
df = pd.DataFrame(data)

# Select the top 10 records
top_10 = df.head(10)

X = list(top_10.iloc[:, 6])
Y = list(top_10.iloc[:, 2])
# Set the figure size to increase width
plt.figure(figsize=(15, 6))  # Adjust the width and height as needed
# Plot the data using bar() method
plt.bar(X, Y, color='g')
plt.title("Top 10 House Rent Analysis")
plt.xlabel("Area of Locality")
plt.ylabel("Rent")

# Show the plot
plt.show()
```

Creating the pie chart for rent analysis

```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file
data = pd.read_csv("House_Rent_Dataset.csv")

# Create a DataFrame
df = pd.DataFrame(data)

# Select the top 10 records
top_10 = df.head(10)

# Replace 'Area of Locality' and 'Rent' with the correct column names from your dataset
# You should use the actual column names from your CSV file
area_locality = top_10['Area Locality']
rent = top_10['Rent']

# Create a pie chart
plt.figure(figsize=(8, 8))  # Set the figure size
plt.pie(rent, labels=area_locality, autopct=lambda p: '{:.0f}'.format(p * sum(rent)/100), startangle=140)
plt.title("Top 10 House Rent Analysis")
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle

# Add a legend in the top left corner
plt.legend(title = "House Rent Analysis",
           loc = "center left",
           bbox_to_anchor = (1,-0.6,1,1))
# Show the pie chart
plt.show()
```

Line graph for the House Rent Analysis

```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file
data = pd.read_csv("House_Rent_Dataset.csv")

# Create a DataFrame
df = pd.DataFrame(data)

# Select the top 12 records
top_10 = df.tail(12)

# Extract the 'Area of Locality' and 'Rent' columns
area_locality = top_10['Area Locality']
rent = top_10['Rent']

# Create a line graph
plt.figure(figsize=(10, 6))  # Set the figure size
plt.plot(area_locality, rent, marker='o', color='b', linestyle='-')
plt.title("Top 12 House Rent Analysis")
plt.xlabel("Area of Locality")
plt.ylabel("Rent")

# Rotate the x-axis labels for better readability (adjust as needed)
plt.xticks(rotation=45)

# Show the line graph
plt.show()
```

Histogram For the House Rent Analysis </br>
A histogram is a visual representation of data distribution, with bins on the x-axis and the frequency of data within each bin on the y-axis. It helps reveal data patterns, central tendencies, and outliers in a dataset.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file
data = pd.read_csv("House_Rent_Dataset.csv")

# Create a DataFrame
df = pd.DataFrame(data)

# Select the top 15 records
top_10 = df.head(15)

# Extract the 'Rent' column
rent = top_10['Rent']

# Create a histogram
plt.figure(figsize=(8, 6))  # Set the figure size
plt.hist(rent, bins=10, color='blue', edgecolor='black')
plt.title("Histogram of Rent for Top 15 Records")
plt.xlabel("Rent")
plt.ylabel("Frequency")

# Show the histogram
plt.show()
```

Scatter Plot For the House Rent Analysis </br>

A scatter plot is a two-dimensional data visualization that displays individual data points as dots on a coordinate plane. It's used to assess the relationship or correlation between two variables, with one variable plotted on the x-axis and the other on the y-axis. Scatter plots are helpful for identifying patterns, trends, clusters, or outliers in data.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file
data = pd.read_csv("House_Rent_Dataset.csv")

# Create a DataFrame
df = pd.DataFrame(data)

# Select the top 10 records
top_30 = df.head(30)

# Extract the 'Area Locality' and 'Rent' columns
area_locality = top_30['Area Locality']
rent = top_30['Rent']

# Create a scatter plot
plt.figure(figsize=(10, 6))  # Set the figure size
plt.scatter(area_locality, rent, color='b', marker='o')
plt.title("Scatter Plot of Rent vs. Area Locality (Top 30 Records)")
plt.xlabel("Area of Locality")
plt.ylabel("Rent")

# Rotate the x-axis labels for better readability (adjust as needed)
plt.xticks(rotation=45)

# Show the scatter plot
plt.show()
```
